```python
import os
volume_path = './ShapeNet/ShapeNetVox32/02691156/879ebdd198cf4aa58f6810e1a2b6aa04/model.binvox'

# if not os.path.exists(directory):
#     os.makedirs(directory)
    
# a = np.load('../ShapeNet/ShapeNetVox32/02691156/879ebdd198cf4aa58f6810e1a2b6aa04')
with open(volume_path, 'rb') as f:
    volume = utils.binvox_rw.read_as_3d_array(f)
    volume = volume.data.astype(np.float32)
    volume = volume.transpose(2,0,1)
    
# volume = np.reshape(a, [32,32,32])
fig = plt.figure()
ax = fig.gca(projection=Axes3D.name)
ax.voxels(volume, facecolors='blue', edgecolor="k", shade=False)
ax.view_init(elev=0, azim=0)
# ax.view_init(elev=90-28.3741324967, azim=-97.4649461834)
plt.axis('off')
# plt.savefig("49/00.png")
# plt.close()
```

```python
R, T = look_at_view_transform(dist=1, elev=-28.3741324967, azim=97.4649461834)
R, T = R.numpy(), T.numpy()
R, T = np.asmatrix(R),np.asmatrix(T)
M = np.zeros([4,4])
M[0:3, 0:3] = R
M[0:3,3] = T
x,y,z = np.nonzero(volume)
voxles = np.ones([len(x),4])
voxles[:,0] = x
voxles[:,1] = y
voxles[:,2] = z
after = np.matmul(voxles,M)
# after = np.matrix.round(after[:,0:3])
after = after[:,0:3]
after = after.astype(np.int32)
after[:,0] = after[:,0]-np.min(after[:,0])
after[:,1] = after[:,1]-np.min(after[:,1])
after[:,2] = after[:,2]-np.min(after[:,2])
after = np.transpose(after)
mask = np.zeros(volume.shape)
mask[after[0],after[1],after[2]] = 1
```

```python
after = np.transpose(after)
mask = np.zeros(volume.shape)
mask[after[0],after[1],after[2]] = 1
volume = mask
# volume = np.reshape(a, [32,32,32])
fig = plt.figure()
ax = fig.gca(projection=Axes3D.name)
ax.voxels(volume, facecolors='blue', edgecolor="k", shade=False)
ax.view_init(elev=0, azim=0)
# ax.view_init(elev=90-28.3741324967, azim=-97.4649461834)
plt.axis('off')
```
